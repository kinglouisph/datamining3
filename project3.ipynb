```python
from datetime import date
import time
```

```python
arr = []
for i in range(len(prices)):
    d = date.fromisoformat(prices.date[i])
    arr.append(time.mktime(d.timetuple()))
```

```python
#fundd = fund
#fund = fund[:]

#takes a while

data = []
funds = {}
for i in range(len(fund)):
    arr = {}
    if i % 5 == 0:
        print(i)
    t = fund['Ticker Symbol'][i]
    d = time.mktime(date.fromisoformat(fund['Period Ending'][i]).timetuple())
    minn=99999999999
    mini=0
    for j in range(len(prices)):
        
        if prices.symbol[j] == t:
            d2 = prices.daydates[j]
            if (d > d2):
                continue
            if d2 - d < minn:
                minn = d2-d
                mini=j
                if minn == 0:
                    break
    arr['open'] = prices['open'][mini]
    arr['low'] = prices['low'][mini]
    arr['high'] = prices['high'][mini]
    arr['volume'] = prices['volume'][mini]
    arr['close'] = prices['close'][mini]
    
    for col in fund.columns[3:]:
        arr[col] = fund[col][i]
        
    if t in funds.keys():
        funds[t].append(arr)
        #   for col in fund.columns[3:]:
            #arr[col] = funds[t][-1][col] - arr[col]
        
   #     funds[t].append(arr)
    else:
        funds[t]=[arr]
```

```python
actuallist = []
for i in funds:
    for j in funds[i]:
        actuallist.append(j)
```

```python
data = pd.DataFrame.from_dict(actuallist)
```

```python
targets = ['close', 'high', 'low', 'volume']
#useless = ['Unnamed: 0', 'Ticker Symbol', 'Period Ending']
train_data = np.array(data2.drop(targets, axis=1))
test_data = np.array(data2[:][targets])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(train_data, test_data, test_size=.2)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import BayesianRidge
from sklearn.linear_model import LinearRegression
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_squared_error
```

```python
poly = PolynomialFeatures(degree=2)
px_train = poly.fit_transform(x_train)
px_test = poly.fit_transform(x_test)
```

```python
px_train.shape
```

```python
mods=[]
for i in range(len(targets)):
    mod = LinearRegression().fit(px_train, y_train.T[i])
    p = mod.predict(px_test)
    print(targets[i])
    print(math.sqrt(mean_squared_error(mod.predict(px_train), y_train.T[i])))
    print(math.sqrt(mean_squared_error(p, y_test.T[i])))
    mods.append(mod)
    
```
